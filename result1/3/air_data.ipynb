```python
import numpy as np
import pandas as pd
```

```python
air = pd.read_csv('7-3.csv')
```

```python
air1 = pd.read_csv('7-4.csv')
```

```python
air = air.iloc[952:1432,:]
air = air.reset_index(drop=True)
```

```python
air_data = pd.DataFrame()

air_data['timestamp(Asia/Seoul)'] = air['timestamp(Asia/Seoul)']
air_data['temp(°C)'] = (air['temp(°C)'] + air1['temp(°C)'])/2
air_data['humid'] = (air['humid'] + air1['humid'])/2
air_data['co2'] = (air['co2'] + air1['co2'])/2
air_data['voc'] = (air['voc'] + air1['voc'])/2
air_data['pm25'] = (air['pm25'] + air1['pm25'])/2
air_data['light'] = (air['light'] + air1['light'])/2
air_data['noise'] = (air['noise'] + air1['noise'])/2
```

```python
air_data.head(-1)
```

```python
air_data['timestamp(Asia/Seoul)'] = pd.to_datetime(air_data['timestamp(Asia/Seoul)']) 

air_data.set_index(air_data['timestamp(Asia/Seoul)'], inplace=True)

air_data = air_data.resample('60T', how={'temp(°C)':np.average,'humid':np.average, 
                            'co2':np.average, 'voc':np.average, 'pm25':np.average, 'light':np.average, 'noise':np.average}).fillna(0)

```
