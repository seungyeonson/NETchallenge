```python
import pandas as pd
import numpy as np
```

```python
ap_data = pd.read_excel("exact_data_9201013_log_7.xlsx")
ap_data = ap_data.drop('datetime',axis = 1)
```

```python
ap_data = ap_data[:240]
```

```python
real_data = pd.read_excel("real_number.xlsx")
```

```python
real_data = real_data[:550]
```

```python
real_data
```

```python
from matplotlib import pyplot as plt
import numpy as np

x = ap_data['date']
y = ap_data['all_num']
y2 = real_data['real_number']

plt.figure(figsize = (15,15))
plt.plot(x,y,'b',label='Fake')
plt.plot(x,y2,'r',label='Real')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('matplotlib sample')
plt.legend(loc='upper right')
plt.show()
```
