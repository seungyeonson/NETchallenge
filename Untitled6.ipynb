```python
ap_data = pd.read_excel('ap.xlsx')
ap_data.head(3)
```

```python
ap_table = ap_data[["AUTHENTICATION_TIME","AP가 있는 건물-층"]]
ap_table.head(3)
```

```python
#ap_table = ap_table.loc[ap_table["AP가 있는 건물-층"]=="명학회관-4층",:]
ap_table = ap_table.loc[(ap_table["AP가 있는 건물-층"]=="명학회관-4층")|(ap_table["AP가 있는 건물-층"]=="명학회관-3층"),:]
ap_table.head(50)
```

```python
ap_table['all_num'] = 1
ap_table = ap_table.drop('AP가 있는 건물-층',1)
ap_table.head(10)
```

```python
ap_table.set_index(['AUTHENTICATION_TIME'], inplace=True, drop=False)
#ap_table.reset_index( inplace= True)
ap_table.head(10)
```
