```python
date_data =pd.DataFrame()
date_data.loc[:,'date_5_6']=pd.Series(pd.date_range("2019-5-1", "2019-7-01", freq="5T"))
```

```python
date_data.head(20)
```

```python
#여기서 00:00:00의 all_num의 의미는 0-5분까지의 사람 숫자로 계산할 것이다.
date_data['all_num']=0
date_data.head(20)
```

```python
import datetime
def round_to_5min(t):
    """ This function rounds a timedelta timestamp to the nearest 5-min mark"""
    t = datetime.datetime(t.year, t.month, t.day, t.hour, t.minute - t.minute%5, 0)  
    return t
```

```python
for row in data_ap.itertuples():
    print(date_data.loc[date_data['date_5_6']==round_to_5min(row.START_USING_TIME)])
    
```

```python
for row in data_ap.itertuples():
    date_data.loc[date_data['date_5_6']==round_to_5min(row.START_USING_TIME),['all_num']]+=1
    
```
