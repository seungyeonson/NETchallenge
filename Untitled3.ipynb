```python
from sklearn.preprocessing import scale
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline

train_Average = train_df['Average'].values.reshape(-1,1)
kmeans = KMeans(n_clusters = 5, init='k-means++', max_iter = 300, random_state = 0).fit(train_Average)
train_Average = pd.DataFrame(data = train_Average, columns = ['Average'])
train_Average['cluster'] = kmeans.labels_
train_Average.head(3)

train_cluster = train_Average.groupby('cluster').count()
print(train_cluster)
```

```python
from sklearn.preprocessing import scale
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

import json

train_df = pd.read_csv('train.csv', sep = ',')
train_df = train_df.dropna()

train_df['Average'] = (train_df['High'] + train_df['Low'])/2

train_Average = train_df['Average'].values.reshape(-1,1)
kmeans = KMeans(n_clusters = 5, init='k-means++', max_iter = 300, random_state = 0).fit(train_Average)
train_Average = pd.DataFrame(data = train_Average, columns = ['Average'])
train_Average['cluster'] = kmeans.labels_

train_df1 = pd.read_csv('train.csv', sep = ',')
train_df1 = train_df.dropna()

train_df1 = train_df1['Average'].values.reshape(-1,1)

score_samples = silhouette_samples(train_df1.data, train_Average['cluster'])
train_Average['silhouette_samples'] = score_samples

average_score = silhouette_score(train_df1.data, train_Average['cluster'])
print('주가 데이터 세트 silhouette Analysis Score:{0:.3f}'.format(average_score))
train_Average.head(3)
train_Average.groupby('cluster')['silhouette_samples'].mean()

# #  dataframe->list
# train_Average = train_Average.values.T.tolist()

# # Print JSON
# print(json.dumps(train_Average, ensure_ascii=False, indent="\t"))

# # Write JSON
# with open('Train_Average.json', 'w', encoding="utf-8") as make_file:
#     json.dump(train_Average, make_file, ensure_ascii=False, indent="\t")
```
