```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ap_df = pd.read_excel('기초특성화과학관1_56.xlsx')
```

```python
ap_df.head(10)
```

```python
#이산화탄소와 사람 수

x = ap_df['date_5_6']
y = ap_df['all_num']
y1 = ap_df['열람실1(이산화탄소)']/30

plt.figure(figsize = (30,15))
plt.plot(x,y)
plt.plot(x,y1)
plt.show()
```

```python
#온도와 사람 수

x = ap_df['date_5_6']
y = ap_df['all_num']
y1 = ap_df['열람실1(온도 ℃)']

plt.figure(figsize = (30,15))
plt.plot(x,y)
plt.plot(x,y1)
plt.show()
```
